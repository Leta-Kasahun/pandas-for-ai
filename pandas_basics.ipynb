<h2 style="color:steelblue;font-family:san-serif">Series<h2>
<h5>Series is A pandas 1-D labeled array that can hold any data type think of like single column in a spreadsheet<h2>

```python
import pandas as pd
data=[100,200,300,400]
series=pd.Series(data)
print(series)

calories={"day 1":1990,"day 2":2009,"day 3":4390,"day 4":1320}
series1=pd.Series(calories)
series1["day 2"]+=989
series1["day 5"]=455
print(series1)
print(series1.loc["day 1"])
```

<h2 style="color:steelblue">DataFrames<h2>
<h5>Dataframe A tabular data structure with rows and columns. similars to an excel spreadsheet.<h5>

```python
import pandas as pd
data={"countries":["Ethiopia","Sudan","Kenya","India","China"],
      "population":[10000000,20000100,30000200, 80000000,90000006]}
df=pd.DataFrame(data)
#adding new column
print(df)
print(df.loc[2])
df["Area (km^2)"]=[2003872,5999478,5883022,4839934,9493033]
print("after adding new column")

print(df)
#adding new row
new_row=pd.DataFrame([{"countries":"Djibouti","population":345388,"Area (km^2)":56489}],index=[5])
print("after adding new row")
df=pd.concat([df,new_row])
print(df)


```
