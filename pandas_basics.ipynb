<h2 style="font-family: sans-serif;">pandas Series</h2>

<p style="font-family: sans-serif; ">
A <strong>pandas Series</strong> is a 1-dimensional labeled array that can hold numbers, text, or Python objects.  
Think of it as a <em>single column in a spreadsheet</em> with an <strong>index</strong>.  
Each column in a DataFrame is a Series and is essential for AI data preparation.
</p>

<ul style="font-family: sans-serif;  line-height:1.5;">
  <li><strong>Index:</strong> Labels each element for easy selection and alignment.</li>
  <li><strong>Vectorized Operations:</strong> Apply arithmetic or logical operations to all elements at once.</li>
  <li><strong>Data Types:</strong> Supports numeric, categorical, and text — useful for ML features.</li>
  <li><strong>NumPy Integration:</strong> Most Series operations use NumPy under the hood for efficiency.</li>
  <li><strong>Selection & Filtering:</strong> Access elements by label or position, or filter by conditions.</li>
  <li><strong>Missing Data:</strong> NaN values are handled naturally — important for cleaning datasets.</li>
</ul>

```python
import pandas as pd
data=[100,200,300,400]
series=pd.Series(data)
print(series)

calories={"day 1":1990,"day 2":2009,"day 3":4390,"day 4":1320}
series1=pd.Series(calories)
series1["day 2"]+=989
series1["day 5"]=455
print(series1)
print(series1.loc["day 1"])
```

<h2>Introducing DataFrames</h2>

<p>
DataFrames are 2-dimensional labeled structures in pandas — like tables in a spreadsheet.  
Mastering them is essential for AI and machine learning workflows. Key manipulations include inspecting data, sorting, subsetting, and adding new columns.
</p>

<ul>
  <li><strong>Introducing DataFrames:</strong> Understand the structure of rows, columns, and indices.</li>
  <li><strong>Inspecting a DataFrame:</strong> Use <code>.head()</code>, <code>.tail()</code>, <code>.info()</code>, and <code>.describe()</code> to explore data.</li>
  <li><strong>Parts of a DataFrame:</strong> Columns (Series), index (row labels), and underlying data.</li>
  <li><strong>Sorting Rows:</strong> Sort by values or index using <code>.sort_values()</code> and <code>.sort_index()</code>.</li>
  <li><strong>Subsetting Columns:</strong> Select one or more columns using <code>df['col']</code> or <code>df[['col1','col2']]</code>.</li>
  <li><strong>Subsetting Rows:</strong> Filter rows by conditions, positions, or categorical variables.</li>
  <li><strong>Adding New Columns:</strong> Create new columns by assignment, calculations, or applying functions.</li>
  <li><strong>Combo-Attack:</strong> Combine multiple manipulations to explore and clean real-world datasets efficiently.</li>
</ul>

```python
import pandas as pd
data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"],
    "Steps": [7000, 8500, 9000, 10000, 7500],
    "Calories": [2200, 2100, 2300, 2500, 2000]
}

df=pd.DataFrame(data)
print(df)
print("selecting only one column")
print(df["Steps"])
print("displaying where the steep is above 8000")
print(df[df["Steps"]>8000])
#inspecting dataframe  to understund the data  before  performing opration  on the  dataframe
print("Instpecting using .head() to return  the first 5 rows")
#to show these we have to upadst to adds two rows
new_row=pd.DataFrame({
    "Day":["Saturday","Sunday"],
    "Steps":[8500,6780],
    "Calories":[2440,2123]
})
#concatnating them 
df=pd.concat([df,new_row],ignore_index=True)

print(df.head())

print("Inspecting using .tail()  to get the last 5 rows")
print(df.tail())
```
