<h2 style="font-family: sans-serif;">pandas Series</h2>

<p style="font-family: sans-serif; ">
A <strong>pandas Series</strong> is a 1-dimensional labeled array that can hold numbers, text, or Python objects.  
Think of it as a <em>single column in a spreadsheet</em> with an <strong>index</strong>.  
Each column in a DataFrame is a Series and is essential for AI data preparation.
</p>

<ul style="font-family: sans-serif;  line-height:1.5;">
  <li><strong>Index:</strong> Labels each element for easy selection and alignment.</li>
  <li><strong>Vectorized Operations:</strong> Apply arithmetic or logical operations to all elements at once.</li>
  <li><strong>Data Types:</strong> Supports numeric, categorical, and text — useful for ML features.</li>
  <li><strong>NumPy Integration:</strong> Most Series operations use NumPy under the hood for efficiency.</li>
  <li><strong>Selection & Filtering:</strong> Access elements by label or position, or filter by conditions.</li>
  <li><strong>Missing Data:</strong> NaN values are handled naturally — important for cleaning datasets.</li>
</ul>

```python
import pandas as pd
data=[100,200,300,400]
series=pd.Series(data)
print(series)

calories={"day 1":1990,"day 2":2009,"day 3":4390,"day 4":1320}
series1=pd.Series(calories)
series1["day 2"]+=989
series1["day 5"]=455
print(series1)
print(series1.loc["day 1"])
```

<h2>Introducing DataFrames</h2>

<p>
DataFrames are 2-dimensional labeled structures in pandas — like tables in a spreadsheet.  
Mastering them is essential for AI and machine learning workflows. Key manipulations include inspecting data, sorting, subsetting, and adding new columns.
</p>

<ul>
  <li><strong>Introducing DataFrames:</strong> Understand the structure of rows, columns, and indices.</li>
  <li><strong>Inspecting a DataFrame:</strong> Use <code>.head()</code>, <code>.tail()</code>, <code>.info()</code>, and <code>.describe()</code> to explore data.</li>
  <li><strong>Parts of a DataFrame:</strong> Columns (Series), index (row labels), and underlying data.</li>
  <li><strong>Sorting Rows:</strong> Sort by values or index using <code>.sort_values()</code> and <code>.sort_index()</code>.</li>
  <li><strong>Subsetting Columns:</strong> Select one or more columns using <code>df['col']</code> or <code>df[['col1','col2']]</code>.</li>
  <li><strong>Subsetting Rows:</strong> Filter rows by conditions, positions, or categorical variables.</li>
  <li><strong>Adding New Columns:</strong> Create new columns by assignment, calculations, or applying functions.</li>
  <li><strong>Combo-Attack:</strong> Combine multiple manipulations to explore and clean real-world datasets efficiently.</li>
</ul>

```python
import pandas as pd
data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"],
    "Steps": [7000, 8500, 9000, 10000, 7500],
    "Calories": [2200, 2100, 2300, 2500, 2000]
}

df=pd.DataFrame(data)
print(df)
print("selecting only one column")
print(df["Steps"])
print("displaying where the steep is above 8000")
print(df[df["Steps"]>8000])
```

<h2>Step 3: DataFrame Parts</h2>

<h5>A pandas DataFrame has key components that define its structure and are essential for data inspection, manipulation, and AI preprocessing.</h5>

<ul>
  <li><strong>.index</strong> – Row labels of the DataFrame. Allows row selection, alignment, and time series labeling.</li>
  <li><strong>.columns</strong> – Column labels of the DataFrame. Enables feature selection, renaming, and descriptive datasets.</li>
  <li><strong>.values</strong> – The underlying NumPy array of data. Needed for fast computations and ML model input.</li>
  <li><strong>.shape</strong> – Tuple of (rows, columns). Quickly understand dataset size for processing.</li>
  <li><strong>.dtypes</strong> – Data type of each column. Ensures correct numeric/categorical handling for analysis or AI.</li>
  <li><strong>.T</strong> – Transpose of the DataFrame (swap rows & columns). Provides alternative view and helps in mathematical operations.</li>
</ul>

```python
print(df.columns)
df.columns=["Day_name","Step_count","Cal_burned"]
print("after column is renamed")
print(df.columns)
print("only values")
print(df.values)
print("only index")
print()
print(df.index)
df.index=[1,2,3,4,5,6,7]
print("After updating the index",df.index)
print("Undestanding clo and row quickly uning .shape",df.shape)
print("datatype ",df.dtypes)
```

<h2>Step 4: Sorting and Subsetting DataFrames</h2>

<h5>
Sorting is used to arrange data in ascending or descending order based on column values. 
It helps find highest, lowest, or chronological patterns in the dataset.
</h5>

<h5>
Subsetting is used to select specific rows or columns from a DataFrame. 
It allows you to focus only on relevant data for analysis or model training.
</h5>

<ul>
  <li><strong>Sorting</strong> – Organizes data using <code>.sort_values()</code> or <code>.sort_index()</code>. Used to rank or order records.</li>
  <li><strong>Subsetting Columns</strong> – Select specific columns using <code>df["col"]</code> or <code>df[["col1","col2"]]</code>. Used to choose features.</li>
  <li><strong>Subsetting Rows</strong> – Filter rows using conditions like <code>df[df["col"] > value]</code>. Used to select specific records.</li>
  <li><strong>Combined Selection</strong> – Use <code>.loc[]</code> or <code>.iloc[]</code> to select rows and columns together.</li>
</ul>

```python
#sorting by rows
df_sorted=df.sort_values(by="Steps")
print("after sorting in aceding by Steps")
print(df_sorted)
#sorting in deceding orders
df_sorted_dec=df.sort_values(by="Steps",ascending=False)
print("After sorting in decending order")
print(df_sorted_dec)

#subneting  means selecting  by columns  oe  or more columns returns  series
df_sebneted=df[["Day","Calories"]]
print("After selecting  day and calories only")
print(df_sebneted)
#subneting  by rows   

df_high_steeps=df[df["Steps"]>7500]
#this nesting  array is inner to  returen the boolens  where  the conditios  ans  true or false  whre is the outer is to  returns  the  filterig data

print("After selecting  highest steps  by  row")
print(df_high_steeps)
#subneting  by multple condition
df_mul_subnet=df[(df["Steps"]>8000) & (df["Calories"]<2500)]
print("after subneting  by multple")
print(df_mul_subnet)
#finltering  some  rowsn usig isin()
weekend=df["Day"].isin(["Saturday","Sunday"])
print("Weekend of week")
print(weekend)


```

<h2>Step 5: Adding New Columns</h2>

<h5>Adding new columns in a pandas DataFrame is used to create new features from existing data. This is important in data analysis and AI preprocessing.</h5>

<ul>
  <li><strong>Direct Assignment</strong> – Create a new column using calculations from existing columns. Used for feature engineering.</li>
  <li><strong>Conditional Columns</strong> – Create True/False columns based on conditions. Used for classification or filtering.</li>
  <li><strong>Using Functions (.apply)</strong> – Apply custom logic to create categorical or derived features.</li>
  <li><strong>Vectorized Operations</strong> – Perform operations on entire columns without loops for better performance.</li>
</ul>

```python
data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Steps": [7000, 8500, 9000, 10000, 7500, 8000, 6700],
    "Calories": [2200, 2100, 2300, 2500, 2000, 2400, 1900]
}

df = pd.DataFrame(data)
print(df)
df["Sum_Steps"]=df["Steps"].sum()
print("After new column is added")
print(df)
df["Calories_per_1000"]=df["Calories"]/(df["Steps"]/1000)
print("After adding using multple columns")
print(df)
```
