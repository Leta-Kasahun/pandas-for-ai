<h2 style="font-family: sans-serif;">pandas Series</h2>

<p style="font-family: sans-serif; ">
A <strong>pandas Series</strong> is a 1-dimensional labeled array that can hold numbers, text, or Python objects.  
Think of it as a <em>single column in a spreadsheet</em> with an <strong>index</strong>.  
Each column in a DataFrame is a Series and is essential for AI data preparation.
</p>

<ul style="font-family: sans-serif;  line-height:1.5;">
  <li><strong>Index:</strong> Labels each element for easy selection and alignment.</li>
  <li><strong>Vectorized Operations:</strong> Apply arithmetic or logical operations to all elements at once.</li>
  <li><strong>Data Types:</strong> Supports numeric, categorical, and text — useful for ML features.</li>
  <li><strong>NumPy Integration:</strong> Most Series operations use NumPy under the hood for efficiency.</li>
  <li><strong>Selection & Filtering:</strong> Access elements by label or position, or filter by conditions.</li>
  <li><strong>Missing Data:</strong> NaN values are handled naturally — important for cleaning datasets.</li>
</ul>

```python
import pandas as pd
data=[100,200,300,400]
series=pd.Series(data)
print(series)

calories={"day 1":1990,"day 2":2009,"day 3":4390,"day 4":1320}
series1=pd.Series(calories)
series1["day 2"]+=989
series1["day 5"]=455
print(series1)
print(series1.loc["day 1"])
```

<h2>Step 5: Adding New Columns</h2>

<h5>Adding new columns in a pandas DataFrame is used to create new features from existing data. This is important in data analysis and AI preprocessing.</h5>

<ul>
  <li><strong>Direct Assignment</strong> – Create a new column using calculations from existing columns. Used for feature engineering.</li>
  <li><strong>Conditional Columns</strong> – Create True/False columns based on conditions. Used for classification or filtering.</li>
  <li><strong>Using Functions (.apply)</strong> – Apply custom logic to create categorical or derived features.</li>
  <li><strong>Vectorized Operations</strong> – Perform operations on entire columns without loops for better performance.</li>
</ul>

```python
data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Steps": [7000, 8500, 9000, 10000, 7500, 8000, 6700],
    "Calories": [2200, 2100, 2300, 2500, 2000, 2400, 1900]
}

df = pd.DataFrame(data)
print(df)
df["Sum_Steps"]=df["Steps"].sum()
print("After new column is added")
print(df)
df["Calories_per_1000"]=df["Calories"]/(df["Steps"]/1000)
print("After adding using multple columns")
print(df)
#implemetnign using  .apply() functios     to adds costomanies columns
def activity_level(steps):
    if steps>=9000:
        return "Very Active"
    elif steps>=7500:
        return "Active"
    else:
        return "Low"
df["Activity_Level"]=df["Steps"].apply(activity_level)    
print("after applying using apply  function ")
print(df)
```

<h2>Step 6: Combo Attack and .assign()</h2>

<h5>
Combo Attack means using multiple DataFrame operations together 
to create new features and prepare data in one logical flow.
It is used when transforming raw data into useful data for analysis or ML.
</h5>

<h5>
Normal Combo Attack modifies the original DataFrame directly 
by adding new columns using df["new_column"] = expression.
It is simple and good for quick transformations.
</h5>

<h5>
.assign() is a pandas method used to add new columns 
without changing the original DataFrame.
It returns a new DataFrame and is useful for clean and professional pipelines.
</h5>

<ul>
  <li><strong>Combo Attack</strong> – Adds multiple columns step by step and modifies the original data.</li>
  <li><strong>.assign()</strong> – Adds multiple columns in one statement and keeps original data safe.</li>
  <li><strong>When to use Combo Attack</strong> – Small projects, quick analysis, simple scripts.</li>
  <li><strong>When to use .assign()</strong> – Clean code, ML preprocessing, professional workflows.</li>
</ul>

```python
data1 = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Steps": [7000, 8500, 9000, 10000, 7500, 8000, 6700],
    "Calories": [2200, 2100, 2300, 2500, 2000, 2400, 1900]
}

df = pd.DataFrame(data1)
df.index=[1,2,3,4,5,6,7]
print("Original Data")
print(df)
print("--------------------------------------------------------------")
print("after updation")
#this compo attack is  implemetig  defferent pandas oparation  togethere
df["Steps_Thousands"]=df["Steps"]/1000
df["Activity_Level"]=df["Steps"].apply(lambda x:"V. Active"if x>=9000 else "Active" if x>=7500 else "Low")
print(df)
#but this is implemete in  veiw  that means  it share the memory and  this is affectes the origninal data  to solves this problem it recommended to implemte using assign
data2 = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Steps": [7000, 8500, 9000, 10000, 7500, 8000, 6700],
    "Calories": [2200, 2100, 2300, 2500, 2000, 2400, 1900]
}
df2=pd.DataFrame(data2)
print("===================================================================================")
print("Original Data")
print(df2)
print("Data manuplation using assing")
new_df=df2.assign(
    Steps_Normalized=df2["Steps"]/df2["Steps"].max(),
    Comulative_Steps=df2["Steps"].cumsum(),
    Step_Ranks=df2["Steps"].rank(ascending=False),
    Calories_Categories=df2["Calories"].apply(lambda x: "Highest" if x>=2400 else "High" if x>=2200 else "Medium" if x>=2000 else "Low"),
    Steps_Percentage=df2["Steps"]/df2["Steps"].sum()
)
print(new_df)
print("-------------------------------------------------------------------")
print("original data after manuplation")
print(df2)
```

<h2>Aggregating DataFrames</h2>

<h5>
Aggregation means summarizing data to extract insights.
It is widely used in data analysis, reporting, and AI preprocessing.
</h5>

<h5>Commonly used operations in real-world projects:</h5>

<ul>
  <li><strong>Mean / Median:</strong> Average or middle values of numeric columns.</li>
  <li><strong>Sum:</strong> Total values, e.g., total steps or calories.</li>
  <li><strong>Min / Max:</strong> Find lowest and highest values (outliers, top performers).</li>
  <li><strong>Count:</strong> Count rows, non-missing values, or category occurrences.</li>
  <li><strong>Cumulative statistics:</strong> Running totals or trends over time.</li>
  <li><strong>Grouped summary (.groupby):</strong> Aggregate data by categories like day, activity level, or type.</li>
  <li><strong>Pivot tables:</strong> Summarize data across multiple dimensions for reports or ML features.</li>
</ul>

<h5>Next steps:</h5>
<p>We will start with the **most common numeric aggregations** (mean, median, sum, min, max) and show examples using your fitness dataset.</p>

```python
import pandas  as pd
data={
    "store":["A","B","C","D","E","F"],
    "product":["X","Y","Z","X","Y","Z"],
    "quantity":[10,15,8,12,25,5],
    "revenue":[100,150,80,120,250,50]
}
df=pd.DataFrame(data)
df.index=[1,2,3,4,5,6]
print(df)
print("Results of Stats summary")
print("--------------------------------------------------------")
print("countes non missing values per column")
print(df.count())
print("min revenue and qty: ",df[["quantity","revenue"]].min())
print("Max revenue and qty: ",df[["quantity","revenue"]].max())
df["cum_rev"]=df["revenue"].cumsum()
print("after communaltive revenu is added")
print(df)
print("----------------------------------------------")
print("Group by opration  to categories based on their columns")
products=df.groupby("product")
print(products.groups)
#usage of gropu  by for  different   data manuplanstion
results=df.groupby("product")["revenue"].sum()
print(results)
#this  is used to in the  real  world  to impltes  their summery  by categories
qty=df.groupby("product")["quantity"].sum()
print("the products  qunatity per products type:")
print(qty)


#agg  this is imprtants this is importants  to apply multple opration at a time
revenue_status=df.groupby("product")["revenue"].agg(["sum","mean","max","min"])
print("after aggregating with agg")
print(revenue_status)
```

<h2>Removing Duplicate Rows (drop_duplicates)</h2>

<h5>
Removing duplicates means cleaning the dataset by eliminating repeated rows or repeated
combinations of values. This is a very common step in data preprocessing because
real-world datasets often contain duplicate records due to data entry errors,
system logging duplication, or repeated imports.
</h5>

<h5>What you need to understand:</h5>

<ul>
  <li><strong>Duplicate rows:</strong> Rows that contain the same values in selected columns.</li>
  <li><strong>subset parameter:</strong> Specifies which columns should be checked for duplicates.</li>
  <li><strong>Data cleaning:</strong> Removing duplicates helps ensure accurate analysis and reporting.</li>
</ul>

<h5>How it works:</h5>
<p>
Pandas scans the specified columns and keeps the first occurrence of each unique
combination while removing the rest.
</p>

<h5>Sample code:</h5>

<pre>
import pandas as pd

sales = pd.DataFrame({
    "store": ["A","A","B","B"],
    "type": ["Super","Super","Discount","Discount"],
    "department": [1,1,2,3]
})

store_types = sales.drop_duplicates(subset=["store","type"])
print(store_types)

Pivot tables are widely used in data analysis, reporting, and data preprocessing.

<ul>

<h5>When pivot tables are used:</h5>

  <li><strong>Reporting and dashboards:</strong> When generating structured summaries for analysis or presentation.</li>
  <li><strong>Data transformation:</strong> When converting long datasets into structured table formats.</li>

<h5>Why pivot tables are important:</h5>

<ul>
  <li><strong>Simplifies complex datasets:</strong> Large and detailed datasets become easier to interpret.</li>
  <li><strong>Reveals patterns:</strong> Helps identify trends and relationships between variables.</li>
  <li><strong>Improves data exploration:</strong> Allows analysts to quickly reorganize data in multiple ways.</li>
  <li><strong>Supports decision-making:</strong> Structured summaries make insights clearer and faster to obtain.</li>
</ul>

<h5>How pivot tables solve problems:</h5>

<ul>
  <li><strong>Organizing data:</strong> Rearranges raw data into structured summaries.</li>
  <li><strong>Reducing redundancy:</strong> Combines repeated records into aggregated values.</li>
  <li><strong>Improving readability:</strong> Makes large datasets easier to analyze and interpret.</li>
</ul>

<br>

<h5>
Pivot tables are one of the most powerful tools in data analysis
because they allow data to be summarized, reorganized, and analyzed
from multiple perspectives with minimal code.
</h5>

```python
import pandas as pd
import pandas as pd

sales = pd.DataFrame({
    "store": ["A","A","B","B","C","C"],
    "product": ["Phone","Laptop","Phone","Laptop","Phone","Laptop"],
    "revenue": [1000,2000,1500,2500,1200,2100]
})

print(sales)

#reorganizing using the pivot table

pivot=pd.pivot_table(
    sales,
    index="store",
    columns="product",
    values="revenue"
)

print("--------------------------------------")
print(pivot)

#pivot table with aggregation
pivot_sum=pd.pivot_table(
    sales,
    index="store",
    columns="product",
    values="revenue",
    aggfunc="sum"
)
print("-------------------------------------")
print(pivot_sum)

pivot_mean = pd.pivot_table(
    sales,
    index="store",
    columns="product",
    values="revenue",
    aggfunc="mean"
)

print(pivot_mean)
```

<h2>List Slicing, loc, and iloc</h2>

<h5>
In data analysis, selecting parts of your dataset is crucial.
Python provides <strong>list slicing</strong> for sequences,
while pandas provides <strong>loc</strong> and <strong>iloc</strong>
for DataFrame selection.
</h5>

<h5>What you need to understand:</h5>

<ul>
  <li><strong>List Slicing:</strong> Extracts elements from a Python list or array using <code>[start:stop:step]</code>.</li>
  <li><strong>loc:</strong> Label-based indexing in pandas, selects rows/columns by names.</li>
  <li><strong>iloc:</strong> Position-based indexing in pandas, selects rows/columns by integer positions.</li>
</ul>

<h5>When to use them:</h5>

<ul>
  <li><strong>List slicing:</strong> When working with Python lists, tuples, or arrays.</li>
  <li><strong>loc:</strong> When row/column labels are meaningful, e.g., dates, IDs, categories.</li>
  <li><strong>iloc:</strong> When row/column positions are more important than labels, e.g., ML pipelines, numeric analysis.</li>
</ul>

<h5>Why they are important:</h5>

<ul>
  <li><strong>List slicing:</strong> Quickly get a subset of data without loops.</li>
  <li><strong>loc:</strong> Safe selection using meaningful labels; avoids mistakes if index is non-numeric or unsorted.</li>
  <li><strong>iloc:</strong> Quick numeric selection; works like Python slicing but on DataFrames.</li>
</ul>

<h5>Problems they solve:</h5>

<ul>
  <li><strong>List slicing:</strong> Avoids writing manual loops for subset extraction.</li>
  <li><strong>loc:</strong> Select specific rows and columns by label; maintain label alignment.</li>
  <li><strong>iloc:</strong> Select specific rows and columns by position; useful when labels are unknown.</li>
</ul>

<h5>Quick tip:</h5>

<ul>
  <li><strong>Python slicing:</strong> [start:stop:step] → exclusive stop.</li>
  <li><strong>iloc:</strong> [start:stop] → like list slicing, exclusive stop.</li>
  <li><strong>loc:</strong> [label1:label2] → inclusive stop.</li>
</ul>

```python
df = pd.DataFrame({
    "store": ["A","B","C","D"],
    "revenue": [1000, 2000, 1500, 2500]
},index=["r1","r2","r3","r4"])
print(df)
#python lists
revenues = df["revenue"].tolist()
print("\nList slicing [1:3]:")
print(revenues[1:3])
print("\nloc: select multiple rows and columns by labels:")
print(df.loc["r1":"r2"])
print("\niloc: select first three rows and both columns:")
print(df.iloc[0:3,0:2])
```
